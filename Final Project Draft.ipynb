```python
import pandas as pd
pd.set_option('display.max_columns', 5000)
melbourne = pd.read_csv("http://data.insideairbnb.com/australia/vic/melbourne/2022-09-09/data/listings.csv.gz")
melbourne.head()
```

```python
melbourne = pd.get_dummies(melbourne, columns=["room_type","host_response_time"], drop_first=True)
melbourne.head()
```

```python
cols = melbourne.columns.tolist()
print(cols)
```

```python
cols1 = [cols[17]]+cols[15:17]+[cols[21]]+[cols[32]]+[cols[38]]+[cols[52]]+[cols[54]]+cols[59:66]+cols[72:79]
print(cols1)
```

```python
melbourne1 = melbourne[cols1]
melbourne1.head()
```

```python
melbourne1.info()
```

```python
melbourne1.fillna({"host_response_rate" : melbourne1["host_response_rate"].mode()[0], "host_acceptance_rate" : melbourne1["host_acceptance_rate"].mode()[0],"host_is_superhost" : melbourne1["host_is_superhost"].mode()[0],"host_listings_count" : melbourne1["host_listings_count"].mode()[0],"review_scores_accuracy" : melbourne1["review_scores_accuracy"].mode()[0],"review_scores_cleanliness" : melbourne1["review_scores_cleanliness"].mode()[0],"review_scores_checkin": melbourne1["review_scores_checkin"].mode()[0],"review_scores_communication" : melbourne1["review_scores_communication"].mode()[0],"review_scores_location" : melbourne1["review_scores_location"].mode()[0],"review_scores_rating" : melbourne1["review_scores_rating"].mode()[0],"review_scores_value" : melbourne1["review_scores_value"].mode()[0],"reviews_per_month" : melbourne1["reviews_per_month"].mode()[0]},inplace = True)
```

```python
melbourne1['price']=melbourne1['price'].str.replace(r'\D+','', regex=True).astype(int)
melbourne1['price']=melbourne1['price']/100
```

```python
melbourne1['host_acceptance_rate']=melbourne1['host_acceptance_rate'].str.replace(r'\D+','', regex=True).astype(int)
```

```python
melbourne1['host_response_rate']=melbourne1['host_response_rate'].str.replace(r'\D+','', regex=True).astype(int)
```

```python
melbourne1.info()
```

```python
y = melbourne1.iloc[:, 0]
x = melbourne1.iloc[:, 1:]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_model = scaler.fit(x)
```

```python
scaled_x = scaler_model.transform(x)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```

## Logistic regression

```python
lr = LogisticRegression()
lr_model = lr.fit(scaled_x, y)
score = cross_val_score(lr_model, scaled_x, y, cv=5)
mean = np.mean(score)
print(mean)
```

## Decision tree - entropy

```python
nums = range(1, 16)
for n in nums:
    dtc_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=n)
    dtc_entropy_model = dtc_entropy.fit(scaled_x, y)
    score = cross_val_score(dtc_entropy_model, scaled_x, y, cv=5)
    mean = np.mean(score)
    print(n, ":", mean)
```

7 : 0.8725047082677904

## Decision tree - gini

```python
nums = range(1, 16)
for n in nums:
    dtc_gini = DecisionTreeClassifier(criterion="gini", max_depth=n)
    dtc_gini_model = dtc_gini.fit(scaled_x, y)
    score = cross_val_score(dtc_gini_model, scaled_x, y, cv=5)
    mean = np.mean(score)
    print(n, ":", mean)
```

5 : 0.8742047958187215

## Random forest - entropy

```python
nums = range(1, 50)
for n in nums:
    rfc_entropy = RandomForestClassifier(criterion="entropy", n_estimators=n, random_state=0)
    rfc_entropy_model = rfc_entropy.fit(scaled_x, y)
    score = cross_val_score(rfc_entropy_model, scaled_x, y, cv=5)
    mean = np.mean(score)
    print(n, ":", mean)
```

## Random forest - gini

```python
nums = range(1, 50)
for n in nums:
    rfc_gini = RandomForestClassifier(criterion="gini", n_estimators=n, random_state=0)
    rfc_gini_model = rfc_gini.fit(scaled_x, y)
    score = cross_val_score(rfc_gini_model, scaled_x, y, cv=5)
    mean = np.mean(score)
    print(n, ":", mean)
```

## neural networks

```python
import itertools
```

### Neural network - relu

```python
nums = range(1, 11) #single hidden layers
for n in nums:
    nnc_r = MLPClassifier(activation="relu", hidden_layer_sizes=n, max_iter=5000, random_state=0)
    scores = cross_val_score(nnc_r, scaled_x, y, cv=5)
    print(n, ":", np.mean(scores))
```

10 : 0.8774399710889507

```python
a = b = range(1, 6) # 2 hidden layer with 1 to 5 nodes each.
nums = list(itertools.product(a,b))
for n in nums:
    nnc_r = MLPClassifier(activation="relu", hidden_layer_sizes=n, max_iter=5000, random_state=0)
    scores = cross_val_score(nnc_r, scaled_x, y, cv=5)
    print(n, ":", np.mean(scores))
```

```python
a = b = c = range(1, 6) # 3 hidden layer with 1 to 5 nodes each
nums = list(itertools.product(a,b,c))
for n in nums:
    nnc_r = MLPClassifier(activation="relu", hidden_layer_sizes=n, max_iter=5000, random_state=0)
    scores = cross_val_score(nnc_r, scaled_x, y, cv=5)
    print(n, ":", np.mean(scores))
```
